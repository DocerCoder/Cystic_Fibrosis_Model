```python
# -------------------------------
# Imports & Setup
# -------------------------------
%pip install scikit-learn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight
from xgboost import XGBClassifier
import shap
import json

shap.initjs()

# -------------------------------
# Load & Merge Patient Data
# -------------------------------
meta = pd.read_json("cftr2_variant_meta.json").T
df = pd.DataFrame(index=meta.index)

sweat = pd.read_json("cftr2_sweat_chloride.json").T.add_prefix("sweat_")
pseudo = pd.read_json("cftr2_pseudomonas_infection.json").T.add_prefix("pseudo_")
pi = pd.read_json("cftr2_pancreatic_insufficiency.json").T.add_prefix("pi_")
lung = pd.read_json("cftr2_lung_function.json").T.add_prefix("lung_")
df = df.join([sweat, pseudo, pi, lung], how="left")

# -------------------------------
# Parse Lung Function
# -------------------------------
df["lung_best_variant"] = df["lung_lung_function_variant"].apply(
    lambda d: d.get("best_fev1") if isinstance(d, dict) else np.nan
)
df["lung_best_two_cf"] = df["lung_lung_function_two_cf"].apply(
    lambda d: d.get("best_fev1") if isinstance(d, dict) else np.nan
)
df["lung_avg_function"] = pd.to_numeric(
    df[["lung_best_variant", "lung_best_two_cf"]].mean(axis=1),
    errors="coerce"
)

# -------------------------------
# Load CFTR Classes
# -------------------------------
with open("cftr2_variant_classes.json", encoding="utf-8") as f:
    class_data = json.load(f)
classes = pd.DataFrame(class_data).T
classes = classes[["inferred_class"]].rename(columns={"inferred_class": "cftr_class"})
df = df.join(classes, how="left")

# -------------------------------
# Feature Engineering
# -------------------------------
df["sweat_log_chloride"] = np.log1p(df["sweat_sweat_chloride_variant"])
df["lung_avg_function"] = df["lung_avg_function"].fillna(df["lung_avg_function"].median())

df["lung_category"] = pd.cut(
    df["lung_avg_function"],
    bins=[0, 40, 70, 100],
    labels=["severe", "moderate", "normal"]
)
df = pd.get_dummies(df, columns=["lung_category"], prefix="lung")

df.drop(columns=[
    "lung_lung_function_variant", 
    "lung_lung_function_two_cf", 
    "lung_best_variant", 
    "lung_best_two_cf"
], inplace=True, errors="ignore")

# -------------------------------
# Final Features & Labels
# -------------------------------
df = df[df["cftr_class"].notnull()]
X = df[[
    "sweat_log_chloride",
    "pseudo_percent_with_infection_variant",
    "pi_percent_with_pi_variant"
] + [col for col in df.columns if col.startswith("lung_")]]
y = df["cftr_class"].astype(str)

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# -------------------------------
# Train/Test Split
# -------------------------------
X_train, X_test, y_train_enc, y_test_enc = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# -------------------------------
# Class Weight Fix
# -------------------------------
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train_enc),
    y=y_train_enc
)
weight_dict = dict(zip(np.unique(y_train_enc), class_weights))
sample_weights = np.array([weight_dict[label] for label in y_train_enc])

# -------------------------------
# Train Classifier
# -------------------------------
param_grid = {
    "max_depth": [3, 5],
    "learning_rate": [0.05, 0.1],
    "n_estimators": [50, 100]
}
grid = GridSearchCV(
    XGBClassifier(eval_metric="mlogloss", random_state=42),
    param_grid, cv=5, verbose=1, n_jobs=-1
)
grid.fit(X_train, y_train_enc, sample_weight=sample_weights)
best_model = grid.best_estimator_

# -------------------------------
# Evaluation
# -------------------------------
y_pred_enc = best_model.predict(X_test)
y_pred = le.inverse_transform(y_pred_enc)
y_test = le.inverse_transform(y_test_enc)

print("📊 Classification Report:")
print(classification_report(y_test, y_pred))
print("\n📊 Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# -------------------------------
# Feature Importance Plot
# -------------------------------
plt.figure(figsize=(8, 5))
plt.barh(X.columns, best_model.feature_importances_, color="indigo")
plt.xlabel("Feature Importance")
plt.title("Top Predictors of CFTR Functional Class")
plt.tight_layout()
plt.show()

# -------------------------------
# SHAP Summary & Force Plot
# -------------------------------
X_test_numeric = X_test.astype(float)
explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test_numeric)

shap.summary_plot(shap_values, X_test_numeric, class_names=le.classes_)

sample_idx = 0
target_label = "Class I"
target_class = list(le.classes_).index(target_label)
sample_shap = shap_values[target_class][sample_idx]

sample_features = X_test_numeric.iloc[sample_idx].values[:sample_shap.shape[0]]
feature_names = X_test_numeric.columns[:sample_shap.shape[0]]

shap.force_plot(
    explainer.expected_value[target_class],
    sample_shap,
    sample_features,
    feature_names=feature_names,
    matplotlib=True
)
```

```python
# -------------------------------MOST IMPORTANT-------------------------------



# 📦 Imports & Setup
# -------------------------------
%pip install scikit-learn imbalanced-learn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedShuffleSplit, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.utils.class_weight import compute_class_weight
from sklearn.impute import SimpleImputer
from xgboost import XGBClassifier
import shap
import json
from imblearn.over_sampling import SMOTE
from collections import Counter

shap.initjs()

# -------------------------------
# 📁 Load & Merge Patient Data
# -------------------------------
meta = pd.read_json("cftr2_variant_meta.json").T
df = pd.DataFrame(index=meta.index)

sweat = pd.read_json("cftr2_sweat_chloride.json").T.add_prefix("sweat_")
pseudo = pd.read_json("cftr2_pseudomonas_infection.json").T.add_prefix("pseudo_")
pi = pd.read_json("cftr2_pancreatic_insufficiency.json").T.add_prefix("pi_")
lung = pd.read_json("cftr2_lung_function.json").T.add_prefix("lung_")
df = df.join([sweat, pseudo, pi, lung], how="left")

# -------------------------------
# 🫁 Parse Lung Function
# -------------------------------
df["lung_best_variant"] = df["lung_lung_function_variant"].apply(
    lambda d: d.get("best_fev1") if isinstance(d, dict) else np.nan
)
df["lung_best_two_cf"] = df["lung_lung_function_two_cf"].apply(
    lambda d: d.get("best_fev1") if isinstance(d, dict) else np.nan
)
df["lung_avg_function"] = pd.to_numeric(
    df[["lung_best_variant", "lung_best_two_cf"]].mean(axis=1),
    errors="coerce"
)

# -------------------------------
# 🧬 Load CFTR Class Labels
# -------------------------------
with open("cftr_variant_classes.json", encoding="utf-8") as f:
    class_data = json.load(f)
classes = pd.DataFrame(class_data).set_index("Variant")
classes = classes[["Reassigned_Class"]].rename(columns={"Reassigned_Class": "cftr_class"})
df = df.join(classes, how="left")

# -------------------------------
# 🛠️ Feature Engineering
# -------------------------------
df["sweat_log_chloride"] = np.log1p(df["sweat_sweat_chloride_variant"])
df["lung_avg_function"] = df["lung_avg_function"].fillna(df["lung_avg_function"].median())
df["lung_category"] = pd.cut(
    df["lung_avg_function"],
    bins=[0, 40, 70, 100],
    labels=["severe", "moderate", "normal"]
)
df = pd.get_dummies(df, columns=["lung_category"], prefix="lung")
df.drop(columns=[
    "lung_lung_function_variant", 
    "lung_lung_function_two_cf", 
    "lung_best_variant", 
    "lung_best_two_cf"
], inplace=True, errors="ignore")

# -------------------------------
# 🧾 Final Features & Labels
# -------------------------------
df = df[df["cftr_class"].notnull()]
X = df[[
    "sweat_log_chloride",
    "pseudo_percent_with_infection_variant",
    "pi_percent_with_pi_variant"
] + [col for col in df.columns if col.startswith("lung_")]]
y = df["cftr_class"].astype(str)

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# -------------------------------
# ✂️ Improved Stratified Split with Class IV emphasis
# -------------------------------
class_iv_idxs = np.where(le.inverse_transform(y_encoded) == "Class IV")[0]
forced_iv_test = class_iv_idxs[:3]  # force up to 3 samples of Class IV
remaining_idxs = np.setdiff1d(np.arange(len(y_encoded)), forced_iv_test)

sss = StratifiedShuffleSplit(n_splits=1, test_size=(len(y_encoded)*0.2 - len(forced_iv_test))/len(remaining_idxs), random_state=42)
for temp_train_idx, temp_test_idx in sss.split(X.iloc[remaining_idxs], y_encoded[remaining_idxs]):
    train_idx = remaining_idxs[temp_train_idx]
    test_idx = np.concatenate([remaining_idxs[temp_test_idx], forced_iv_test])

X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
y_train_enc, y_test_enc = y_encoded[train_idx], y_encoded[test_idx]

# 🧪 Confirm Class IV inclusion
print("🧬 Unique CFTR classes in test set:")
print(sorted(set(le.inverse_transform(y_test_enc))))

# -------------------------------
# 🧼 Imputation
# -------------------------------
imputer = SimpleImputer(strategy="median")
X_train_imputed = imputer.fit_transform(X_train)

# -------------------------------
# 🔄 SMOTE Resampling
# -------------------------------
y_train_labels = le.inverse_transform(y_train_enc)
smote = SMOTE(
    sampling_strategy={
        'Class II': 100,
        'Class III': 80,
        'Class IV': 50,
        'Class V': 100,
        'Likely non-CF': 60,
        'Uncertain': 200
    },
    k_neighbors=1,
    random_state=42
)
X_resampled, y_resampled_labels = smote.fit_resample(X_train_imputed, y_train_labels)
print("✅ Resampled class distribution:", Counter(y_resampled_labels))

# -------------------------------
# 🏋️ Label Encoding & Class Weights
# -------------------------------
y_resampled_enc = le.transform(y_resampled_labels)
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_resampled_enc),
    y=y_resampled_enc
)
weight_dict = dict(zip(np.unique(y_resampled_enc), class_weights))
sample_weights = np.array([weight_dict[label] for label in y_resampled_enc])

# -------------------------------
# 🧠 Train Classifier
# -------------------------------
param_grid = {
    "max_depth": [3, 5],
    "learning_rate": [0.05, 0.1],
    "n_estimators": [50, 100]
}
grid = GridSearchCV(
    XGBClassifier(eval_metric="mlogloss", random_state=42),
    param_grid, cv=5, verbose=1, n_jobs=-1
)
grid.fit(X_resampled, y_resampled_enc, sample_weight=sample_weights)
best_model = grid.best_estimator_

# -------------------------------
# 🧪 Evaluation
# -------------------------------
X_test_imputed = imputer.transform(X_test)
y_pred_enc = best_model.predict(X_test_imputed)
y_pred = le.inverse_transform(y_pred_enc)
y_test = le.inverse_transform(y_test_enc)

print("📊 Classification Report:")
print(classification_report(y_test, y_pred))
print("\n📊 Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# -------------------------------
# 📈 Feature Importance Plot
# -------------------------------
importances = best_model.feature_importances_
feature_names = X.columns[:len(importances)]

plt.figure(figsize=(8, 5))
plt.barh(feature_names, importances, color="indigo")
plt.xlabel("Feature Importance")
plt.title("Top Predictors of CFTR Functional Class")
plt.tight_layout()
plt.show()

# -------------------------------
# 🔍 SHAP Summary & Force Plot
# -------------------------------
actual_num_features = best_model.feature_importances_.shape[0]
X_test_numeric = pd.DataFrame(X_test_imputed, columns=X.columns[:actual_num_features])

explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test_numeric)

if isinstance(shap_values, np.ndarray) and shap_values.ndim == 3:
    shap_values = shap_values[:, :, :X_test_numeric.shape[1]]

shap.summary_plot(shap_values, X_test_numeric, class_names=le.classes_)

# Force plot for sample
sample_idx = 0
target_label = "Class I"

if isinstance(shap_values, list):
    target_class = list(le.classes_).index(target_label)
    sample_shap = shap_values[target_class][sample_idx]
    expected_value = explainer.expected_value[target_class]
elif shap_values.ndim == 3:
    target_class = list(le.classes_).index(target_label)
    sample_shap = shap_values[sample_idx, target_class]
    expected_value = explainer.expected_value[target_class]
else:
    sample_shap = shap_values[sample_idx]
    expected_value = explainer.expected_value

sample_features = X_test_numeric.iloc[sample_idx].values[:sample_shap.shape[0]]
feature_names = X_test_numeric.columns[:sample_shap.shape[0]]

shap.force_plot(
    expected_value,
    sample_shap,
    sample_features,
    feature_names=feature_names,
    matplotlib=True
)
print(y.value_counts())
```
