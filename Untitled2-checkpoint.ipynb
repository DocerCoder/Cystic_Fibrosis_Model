```python
import os
os.environ["SCIPY_ARRAY_API"] = "1"

%pip install imbalanced-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE
import shap
import json


# -------------------------------
# Load & Prepare Data
# -------------------------------
meta = pd.read_json("cftr2_variant_meta.json").T
df = pd.DataFrame(index=meta.index)

sweat = pd.read_json("cftr2_sweat_chloride.json").T.add_prefix("sweat_")
pseudo = pd.read_json("cftr2_pseudomonas_infection.json").T.add_prefix("pseudo_")
pi = pd.read_json("cftr2_pancreatic_insufficiency.json").T.add_prefix("pi_")
lung = pd.read_json("cftr2_lung_function.json").T.add_prefix("lung_")
df = df.join([sweat, pseudo, pi, lung], how="left")

df["lung_best_variant"] = df["lung_lung_function_variant"].apply(
    lambda d: d.get("best_fev1") if isinstance(d, dict) else np.nan
)
df["lung_best_two_cf"] = df["lung_lung_function_two_cf"].apply(
    lambda d: d.get("best_fev1") if isinstance(d, dict) else np.nan
)
df["lung_avg_function"] = pd.to_numeric(
    df[["lung_best_variant", "lung_best_two_cf"]].mean(axis=1), errors="coerce"
)

with open("cftr2_variant_classes.json", encoding="utf-8") as f:
    class_data = json.load(f)
classes = pd.DataFrame(class_data).T
classes = classes[["inferred_class"]].rename(columns={"inferred_class": "cftr_class"})
df = df.join(classes, how="left")

df["sweat_log_chloride"] = np.log1p(df["sweat_sweat_chloride_variant"])
df["lung_avg_function"] = df["lung_avg_function"].fillna(df["lung_avg_function"].median())

df["lung_category"] = pd.cut(
    df["lung_avg_function"], bins=[0, 40, 70, 100], labels=["severe", "moderate", "normal"]
)
df = pd.get_dummies(df, columns=["lung_category"], prefix="lung", dtype=float)

df.drop(columns=[
    "lung_lung_function_variant", 
    "lung_lung_function_two_cf", 
    "lung_best_variant", 
    "lung_best_two_cf"
], inplace=True, errors="ignore")

# -------------------------------
# Target & Features
# -------------------------------
df = df[df["cftr_class"].notnull()]
# Select only valid columns for X (exclude lung_avg_function if all NaN)
lung_cols = [col for col in df.columns if col.startswith("lung_") and col != "lung_avg_function"]
X = df[[
    "sweat_log_chloride",
    "pseudo_percent_with_infection_variant",
    "pi_percent_with_pi_variant"
] + lung_cols].copy()

# Convert boolean columns to float
for col in ["lung_severe", "lung_moderate", "lung_normal"]:
    if col in X.columns:
        X[col] = X[col].astype(float)

y = df["cftr_class"].astype(str)

# -------------------------------
# Label Encoding & SMOTE
# -------------------------------
le = LabelEncoder()
y_encoded = le.fit_transform(y)

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y_encoded)

# -------------------------------
# Train/Test Split
# -------------------------------
X_train, X_test, y_train_enc, y_test_enc = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

# -------------------------------
# Model Training
# -------------------------------
param_grid = {
    "max_depth": [3, 5],
    "learning_rate": [0.05, 0.1],
    "n_estimators": [50, 100]
}
grid = GridSearchCV(
    XGBClassifier(eval_metric="mlogloss", random_state=42),
    param_grid, cv=5, verbose=1, n_jobs=-1
)
grid.fit(X_train, y_train_enc)
best_model = grid.best_estimator_

# -------------------------------
# Evaluation
# -------------------------------
y_pred_enc = best_model.predict(X_test)
y_pred = le.inverse_transform(y_pred_enc)
y_test = le.inverse_transform(y_test_enc)

print("📊 Classification Report:")
print(classification_report(y_test, y_pred))
print("\n📊 Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# -------------------------------
# Feature Importance
# -------------------------------
plt.figure(figsize=(8, 5))
plt.barh(X.columns, best_model.feature_importances_, color="mediumseagreen")
explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test.astype(float))

# SHAP summary plot for multiclass
shap.summary_plot(shap_values, X_test.astype(float), class_names=le.classes_)

sample_idx = 0
target_label = "Class I"
target_class = list(le.classes_).index(target_label)

# For multiclass, shap_values is [n_classes, n_samples, n_features]
sample_shap = shap_values[target_class][sample_idx]
sample_features = X_test.iloc[sample_idx].values
feature_names = X_test.columns

# Replace NaN in sample_features with 0 for force plot
sample_features = np.nan_to_num(sample_features)

shap.force_plot(
    explainer.expected_value[target_class],
    sample_shap,
    sample_features,
    feature_names=feature_names,
    matplotlib=True
)

shap.force_plot(
    explainer.expected_value[target_class],
    sample_shap,
    sample_features,
    feature_names=feature_names,
    matplotlib=True
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBClassifier
import shap
import json

# -------------------------------
# SHAP Initialization
# -------------------------------
shap.initjs()

# -------------------------------
# Load Data
# -------------------------------
meta = pd.read_json("cftr2_variant_meta.json").T
df = pd.DataFrame(index=meta.index)

sweat = pd.read_json("cftr2_sweat_chloride.json").T.add_prefix("sweat_")
pseudo = pd.read_json("cftr2_pseudomonas_infection.json").T.add_prefix("pseudo_")
pi = pd.read_json("cftr2_pancreatic_insufficiency.json").T.add_prefix("pi_")
lung = pd.read_json("cftr2_lung_function.json").T.add_prefix("lung_")
df = df.join([sweat, pseudo, pi, lung], how="left")

# -------------------------------
# Parse Lung Function
# -------------------------------
df["lung_best_variant"] = df["lung_lung_function_variant"].apply(
    lambda d: d.get("best_fev1") if isinstance(d, dict) else np.nan
)
df["lung_best_two_cf"] = df["lung_lung_function_two_cf"].apply(
    lambda d: d.get("best_fev1") if isinstance(d, dict) else np.nan
)
df["lung_avg_function"] = pd.to_numeric(
    df[["lung_best_variant", "lung_best_two_cf"]].mean(axis=1),
    errors="coerce"
)

# -------------------------------
# Load CFTR Classes
# -------------------------------
with open("cftr2_variant_classes.json", encoding="utf-8") as f:
    class_data = json.load(f)
classes = pd.DataFrame(class_data).T
classes = classes[["inferred_class"]].rename(columns={"inferred_class": "cftr_class"})
df = df.join(classes, how="left")

# -------------------------------
# Feature Engineering
# -------------------------------
df["sweat_log_chloride"] = np.log1p(df["sweat_sweat_chloride_variant"])
df["lung_avg_function"] = df["lung_avg_function"].fillna(df["lung_avg_function"].median())

df["lung_category"] = pd.cut(
    df["lung_avg_function"],
    bins=[0, 40, 70, 100],
    labels=["severe", "moderate", "normal"]
)
df = pd.get_dummies(df, columns=["lung_category"], prefix="lung")

df.drop(columns=[
    "lung_lung_function_variant", 
    "lung_lung_function_two_cf", 
    "lung_best_variant", 
    "lung_best_two_cf"
], inplace=True, errors="ignore")

# -------------------------------
# Target and Features
# -------------------------------
df = df[df["cftr_class"].notnull()]
X = df[[
    "sweat_log_chloride",
    "pseudo_percent_with_infection_variant",
    "pi_percent_with_pi_variant"
] + [col for col in df.columns if col.startswith("lung_")]]
y = df["cftr_class"].astype(str)

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# -------------------------------
# Train/Test Split
# -------------------------------
X_train, X_test, y_train_enc, y_test_enc = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

# -------------------------------
# Model Training
# -------------------------------
param_grid = {
    "max_depth": [3, 5],
    "learning_rate": [0.05, 0.1],
    "n_estimators": [50, 100]
}
grid = GridSearchCV(
    XGBClassifier(eval_metric="mlogloss", random_state=42),
    param_grid, cv=5, verbose=1, n_jobs=-1
)
grid.fit(X_train, y_train_enc)
best_model = grid.best_estimator_

# -------------------------------
# Predictions & Evaluation
# -------------------------------
y_pred_enc = best_model.predict(X_test)
y_pred = le.inverse_transform(y_pred_enc)
y_test = le.inverse_transform(y_test_enc)

print("📊 Classification Report:")
print(classification_report(y_test, y_pred))

print("\n📊 Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# -------------------------------
# Feature Importance
# -------------------------------
plt.figure(figsize=(8, 5))
plt.barh(X.columns, best_model.feature_importances_, color="indigo")
plt.xlabel("Feature Importance")
plt.title("Top Predictors of CFTR Functional Class")
plt.tight_layout()
plt.show()

# -------------------------------
# SHAP Summary & Force Plot — Final Fix
# -------------------------------

# 🔧 Convert test features to float to avoid dtype errors
X_test_numeric = X_test.astype(float)

# 🔧 Create SHAP explainer safely without passing data directly
explainer = shap.TreeExplainer(best_model)
shap_values = explainer.shap_values(X_test_numeric)

# 📊 Summary plot across all classes
shap.summary_plot(shap_values, X_test_numeric, class_names=le.classes_)

# 🔍 Force plot for a specific sample and class
sample_idx = 0
target_label = "Class I"  # Change this to another class label if needed
target_class = list(le.classes_).index(target_label)

# ✅ Extract values safely
sample_shap = shap_values[target_class][sample_idx]

# Ensure the number of features matches the number of SHAP values
num_features = sample_shap.shape[0]
sample_features = X_test_numeric.iloc[sample_idx].values[:num_features]
feature_names = X_test_numeric.columns[:num_features]

# ✅ Render force plot with feature alignment
shap.force_plot(
    explainer.expected_value[target_class],
    sample_shap,
    sample_features,
    feature_names=feature_names,
    matplotlib=True
)
```
